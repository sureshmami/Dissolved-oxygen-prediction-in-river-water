```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
pd.set_option('display.max_rows',100)
pd.set_option('display.max_columns',None)

import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('train.csv')
# test = pd.read_csv('test.csv')
```

```python
train.drop('Id',axis=1,inplace=True)
```

```python
print('percetage of missing values:\n',round(train.isna().sum()/len(train)*100,2))
```

```python
null_values = list(train.columns[train.isna().sum() > 100])
null_values
```

```python
train.drop(null_values,axis=1,inplace=True)
```

```python
print('columns with missing values:', (train.isna().sum(axis=1) != 0).sum())
print('columns with missing values:',(train.isna().sum(axis= 0 ) != 0).sum())
```
